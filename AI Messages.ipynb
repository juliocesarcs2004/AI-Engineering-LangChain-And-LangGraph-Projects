## AI Messages

```python
%load_ext dotenv
%dotenv
```

```python
from langchain_openai.chat_models import ChatOpenAI

from langchain_core.messages import HumanMessage, AIMessage
```

```python
chat = ChatOpenAI(model_name = 'gpt-4',
                  seed = 365,
                  temperature = 0,
                  max_tokens = 100)
```

```python
message_h_dog = HumanMessage(content = ''' I've recently adopted a dog. Can you suggest some dog names?''')
message_ai_dog = AIMessage(content = ''' Oh, absolutely. Because I'm sure your dog is just waiting to be named by a chatbot. 
How about "Bark Twain" if it's a literary hound, or "Sir Wag-a-lot" if it's got a knightly vibe? Maybe "Bark Zuckerberg" 
if it's into social media, or "Bark Obama" if it's a political pooch. ''')

message_h_cat = HumanMessage(content = ''' I've recently adopted a cat. Can you suggest some cat names?''')
message_ai_cat = AIMessage(content = ''' Oh, absolutely. Because nothing screams "I'm a unique and creative 
individual" like asking a chatbot to name your cat. How about "Furry McFurFace", "Sir Meowsalot", or "Purrlock Holmes"? 
Or if you're into pop culture, maybe "Catrick Swayze" or "Purrcey Jackson". You're welcome. ''')

message_h_fish = HumanMessage(content = ''' I've recently adopted a fish. Can you suggest some fish names?''')
```

```python
response = chat.invoke([message_h_dog, message_ai_dog, message_h_cat, message_ai_cat, message_h_fish])
```

```python
print(response.content)
```
